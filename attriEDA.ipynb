```python
df=pd.DataFrame(data)
df
```

```python
# Drop Unwanted Columns
df.drop(columns=["EmployeeCount","Over18","EmployeeNumber","StandardHours"],inplace=True)
```

```python
df.info()
```

```python
# Select only numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

outlier_summary = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Find outlier values
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    
    outlier_summary[col] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outlier Count': outliers.count()
    }

# Convert to DataFrame for better display
outlier_df = pd.DataFrame(outlier_summary).T
print(outlier_df)
```

```python
# EDA- KEY FINDINGS
#1. Employees with comparatively lower performance score had the highest attrition rates.
# Numerical VS Numerical-  T-Test
import pandas as pd
from scipy.stats import ttest_ind

# 1. Encode Attrition to numeric if not already done
df['Attrition_num'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# 2. Define low and high performance groups
low_perf = df[df['PerformanceRating'] == 3]['Attrition_num']
high_perf = df[df['PerformanceRating'] == 4]['Attrition_num']

# 3. Perform independent t-test (Welch’s)
t_stat, p_val = ttest_ind(low_perf, high_perf, equal_var=False)

# 4. Show results

print("T-statistic:", t_stat)
print("p-value:", p_val)

# 5. Interpretation

if p_val < 0.05:
    print("✅ Significant difference: Lower performance scores have different attrition rates.")
else:
    print("❌ No significant difference in attrition rates between the two groups.")

```

```python
#2. Overtime workers had a higher tendency to leave compared to non-overtime employees.

from scipy.stats import chi2_contingency

# Create contingency table
contingency_table = pd.crosstab(df['OverTime'], df['Attrition'])

# Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Contingency Table:\n", contingency_table)
print("\nChi-Square Statistic:", chi2)
print("p-value:", p)

# Interpret result
if p <= 0.05:
    print("\n✅ Significant association: Overtime is related to attrition.")
else:
    print("\n❌ No significant association: Overtime is not related to attrition.")
```

```python
# 3. Lower salary brackets (<5000 monthly income) correlated with increased attrition.

from scipy.stats import chi2_contingency

# Create a new column for income category
df['IncomeCategory'] = df['MonthlyIncome'].apply(lambda x: '<5000' if x < 5000 else '>=5000')

# Create contingency table
contingency_table = pd.crosstab(df['IncomeCategory'], df['Attrition'])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Contingency Table:\n", contingency_table)
print("\nChi-Square Statistic:", chi2)
print("p-value:", p)

# Interpretation
if p <= 0.05:
    print("\n✅ Significant association: Salary category is related to attrition.")
else:
    print("\n❌ No significant association: Salary category is not related to attrition.")
```

```python
#4. Employees far from home (>15 miles) were significantly more likely to resign

from scipy.stats import chi2_contingency

# Create distance category
df['FarFromHome'] = df['DistanceFromHome'].apply(lambda x: 'Far' if x > 15 else 'Near')

# Contingency table
contingency_table = pd.crosstab(df['FarFromHome'], df['Attrition'])

# Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2)
print("p-value:", p)
print("\nContingency Table:\n", contingency_table)

# Interpretation
if p < 0.05:
    print("✅ Significant relationship: Distance from home is linked to attrition.")
else:
    print("❌ No significant relationship: Distance from home is not linked to attrition.")
```

```python
#5. Employees with higher job satisfaction had lower attrition rate

from scipy.stats import ttest_ind

# Convert Attrition to numeric (Yes=1, No=0)
df['Attrition_num'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Split into groups
group_yes = df[df['Attrition'] == 'Yes']['JobSatisfaction']
group_no = df[df['Attrition'] == 'No']['JobSatisfaction']

# Perform t-test
t_stat, p_val = ttest_ind(group_yes, group_no, nan_policy='omit')

print("t-statistic:", t_stat)
print("p-value:", p_val)

# Interpretation
if p_val < 0.05:
    print("✅ Significant difference: JobSatisfaction differs between attrition groups.")
else:
    print("❌ No significant difference: JobSatisfaction is similar between groups.")
```

```python
#6. Does education level influence attrition?

from scipy.stats import chi2_contingency

# Contingency table: Education vs Attrition
contingency_table = pd.crosstab(df['Education'], df['Attrition'])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2)
print("p-value:", p)
print("\nObserved Frequencies Table:\n", contingency_table)

# Interpretation
alpha = 0.05
if p < alpha:
    print("\n✅ Conclusion: There is a statistically significant association between Education level and Attrition.")
else:
    print("\n❌ Conclusion: No statistically significant association between Education level and Attrition.")
```
